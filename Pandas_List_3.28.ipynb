### 리스트 수정하는 법

```python
customerData1 = pd.read_csv("../dataset/kopo_customerdata.csv") #리스트 출력한 것을 새로운 변수에 담는다.
```

```python
customerData1.head(2) #헤드값만 출력해서 리스트가 맞는지 확인한다.
```

```python
customerData1.dtypes # 리스트의 타입이 무엇인지 알아본다.
```

```python
customerData1 = customerData1[(customerData1.GENDER == "Male") & (customerData1.EMAIL > 0)] # 리스트를 수정해서 다시 변수에 넣는 방법.
```

```python
customerData1.head(100) #잘 되었는지 확인해본다. 리스트 양이 많기 때문에 헤드를 이용 100줄만 출력해본다.
```

```python
targetGender = "Female"
```

```python
targetGender = targetGender.upper()
```

```python
targetEmail = 0
```

```python
customerData1 = customerData1[(customerData1.GENDER.str.upper() == targetGender) & (customerData1.EMAIL > targetEmail)]
```
